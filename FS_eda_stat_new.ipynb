```python
# categorical variables in a dataframe
p_cat = df[['site', 'hhedu', 'hhhsex', 'hhethnic', 'povline', 'windex3', 'windex5', 'FS_score']]
p_cat
```

```python
# columns for which pivot tables will be created
pivot_columns = ['site', 'hhedu', 'hhhsex', 'hhethnic', 'povline', 'windex3', 'windex5']

# create pivot tables iteratively for 2 columns
for i in range(len(pivot_columns)):
    for j in range(i + 1, len(pivot_columns)):
        col1 = pivot_columns[i]
        col2 = pivot_columns[j]
        
        print(f"--- Pivot Table for {col1} and {col2} ---")
        pivot_table2 = pd.pivot_table(p_cat, index=col1, columns=col2, values='FS_score')
        print(pivot_table2)
        print("\n")
```

```python
# columns for which pivot tables will be created
pivot_col = ['site', 'hhedu', 'hhhsex', 'hhethnic', 'povline', 'windex3', 'windex5']

# Create pivot tables iteratively for 3 columns
for i in range(len(pivot_col)):
    for j in range(i + 1, len(pivot_col)):
        for k in range(j + 1, len(pivot_col)):
            col1 = pivot_col[i]
            col2 = pivot_col[j]
            col3 = pivot_col[k]
            
            print(f"--- Pivot Table for {col1}, {col2}, and {col3} ---")
            pivot_table3 = pd.pivot_table(df, index=[col1, col2], columns=col3, values='FS_score')
            print(pivot_table3)
            print("\n")
```

```python
# Columns for which pivot tables will be created
pivot_columns = ['site', 'hhedu', 'hhhsex', 'hhethnic', 'povline', 'windex3', 'windex5']

# Create pivot tables iteratively for 3 columns
for i in range(len(pivot_columns)):
    for j in range(i + 1, len(pivot_columns)):
        for k in range(j + 1, len(pivot_columns)):
            col1 = pivot_columns[i]
            col2 = pivot_columns[j]
            col3 = pivot_columns[k]
            
            print(f"--- Pivot Table for {col1}, {col2}, and {col3} ---")
            pivot_table = pd.pivot_table(df, index=[col1, col2], columns=col3, values='FS_score')
            print(pivot_table)
            
            # Create a bar plot
            plt.figure(figsize=(4, 4))
            sns.heatmap(pivot_table, annot=True, cmap='viridis')
            plt.title(f"Pivot Table for {col1}, {col2}, and {col3}")
            plt.tight_layout()

            # Save the plot as an image
            image_file_name = f'pivot_{col1}_{col2}_{col3}.png'
            plt.savefig(image_file_name)
            plt.close()

print("Pivot table images saved.")
```
